```python
from pathlib import Path  #core python module
import win32com.client  #pip install pywin32
import os
from email import utils
import datetime
import re

# Create output folder
output_dir = Path.cwd() / "Output"
output_dir.mkdir(parents=True, exist_ok=True)

# Connect to outlook
outlook = win32com.client.Dispatch("Outlook.Application").GetNamespace("MAPI")

# Connect to folder
b=int(input("Please select folder you want to download attachments from:\n 1. Inbox \n 2. Sent Items \n 3. Reminders \n 4. To Do List \n 5. Contacts\n"))
try:        
    if(b==1):
        i=6
    elif(b==2):
        i=5
    elif(b==3):
        i=14
    elif(b==4):
        i=28
    elif(b==5):
        i=10
except:
       print("Please enter a valid option!")

# Get messages
option = outlook.GetDefaultFolder(i)
messages = option.Items

saving=int(input("How do you want to save the attachments? \n 1. By Sender \n 2. By Subject \n 3. By Date"))
if (saving==3):
    startdate=input("dd/mm/yyyy")
    startdate = datetime.datetime.strptime(startdate, "%d-%m-%Y")
    enddate=input("dd/mm/yyyy")
    enddate = datetime.datetime.strptime(enddate, "%d-%m-%Y")
for message in messages:
    try:
        sender=message.sender
        subject=message.subject
        attachments=message.Attachments
        body = message.body
        date=message.senton.date()
    except:
        print(sender," is empty")


    # Create separate folder for each message
    if (saving==1):
        target_folder = output_dir / str(sender)
        target_folder.mkdir(parents=True, exist_ok=True)
    elif(saving==2):
        subject=re.sub(r'[^\w\s]', '', subject)
        subject=re.sub(r'[0-9]', '', subject)
        target_folder = output_dir / str(subject)
        target_folder.mkdir(parents=True, exist_ok=True)
    elif(saving==3):
        target_folder = output_dir / str(date>startdate)
        target_folder.mkdir(parents=True, exist_ok=True)
    # Write body to text file
    #try:
    #    Path(target_folder / "EMAIL_BODY.txt").write_text(str(body))
    #except:
    #    pass

    # Save attachments
    for attachment in attachments:
        try:
            print("Saving attachments .....")
            attachment.SaveAsFile(target_folder / str(attachment))
        except:
            print('Something went wrong Saving attachments .....')
```
